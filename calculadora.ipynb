```python
def loop():
##comecei definindo uma função pois queria que o código reiniciasse pelo menos uma vez

  print ("Boas vindas a minha calculadora, por gentileza utilize pontinho enves de virgula para numeros decimais ex: 5.5")
  ##mensagem de boas vindas

  valor_primeiro = float (input("\nDigite o valor de primeiro valor do calculo: "))
  valor_segundo = float (input("\nDigite o valor de segundo valor do calculo: "))
  ##recebe os valores que será usado como base no cálculo


  opcao = int (input("\nDigite o numero do calculo que deseja realizar: \n1- Adição \n2- Subtração \n3- Multiplicação \n4- Divisão \n \n"))
  ##recebe a opção de cálculo desejado

## aqui eu quis tester a estrutura de condicional usando if/elif/else, é onde será feito a leitura da opção e com base nisso executará o cálculo desejado
  if opcao == 1:
   adicao = valor_primeiro + valor_segundo
   print("\n", valor_primeiro, " + ", valor_segundo, " = ", adicao,"\n")

  elif opcao == 2:
    subtracao= valor_primeiro - valor_segundo
    print("\n", valor_primeiro, " - ", valor_segundo, " = ", subtracao, "\n")

  elif opcao == 3:
    multiplicacao = valor_primeiro * valor_segundo
    print("\n", valor_primeiro, " x ", valor_segundo, " = ", multiplicacao, "\n")

  elif opcao == 4:
    divisao = valor_primeiro / valor_segundo
    print("\n", valor_primeiro, " / ", valor_segundo, " = ", divisao, "\n")

  else:
    print("\nOpção invalida")
    return
    ##fim da função

loop()
##executa a função criada anteriormente

##aqui criei outra condição tambem com if/elif/else para para o código rodar mais uma vez, queria que o programa me perguntasse ao fim se eu iria recomeçar ou finalizar
sim_não = (input("\nDeseja recomeçar? \nDigite sim ou não: "))

if sim_não == "sim":
    loop()

elif sim_não =="não":
    print("\nObrigado por utilizar o programa, finalizamos aqui.")

else:
    print("\nOpção invalida, programa finalizado.")


print("Fim :)")
##mensagem de finalização do programa









```
